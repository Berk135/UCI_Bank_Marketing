# BANK MARKETING

In this project, I will look at the <i>Bank Marketing Data Set</i> from the UCI repository. 

<h2> 1. Data Set </h2>

First, I will import libraries we will need. Then, I will import the data set using the URL for the relevant UCI archive. If this doesn't work, consider importing the data set locally from your computer.

```python
import pandas as pd
import requests
import zipfile
from io import BytesIO
```

```python
#Get the zipfile
url = "https://archive.ics.uci.edu/ml/machine-learning-databases/00222/bank-additional.zip"
results = requests.get(url)
z = zipfile.ZipFile(BytesIO(results.content))
data = z.open("bank-additional/bank-additional-full.csv")

#Now, we can read in the data
df = pd.read_csv(data, sep = ";")
z.close()
```
