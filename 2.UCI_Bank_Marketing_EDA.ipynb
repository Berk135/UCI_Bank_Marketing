<h2> 2. Exploratory Data Analysis </h2>

Now, let's do Exploratory Data Analysis (EDA) to check the data and create data visuals. First, the csv file saved in the first notebbok has to be loaded.


```python
#Importing required libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats

import warnings
warnings.filterwarnings("ignore")
```

```python
df = pd.read_csv("./bank_marketing.csv")
print(df.shape)
df.head()
```

<h3> 2.1. Summary Statistics </h3>
In this step, I will get the summary statistics for the data set. Yet, first I will check if there are any missing values.

```python
#check for missing values
df.isnull().sum()
```

We conclude that there are no missing values. 

Now, check if there are duplicate rows.

```python
df[df.duplicated()]
```

We see that there are 12 duplicate rows and we will drop them.

```python
df = df.drop_duplicates()
df.shape
```

Now, let's check the data type and summary statistics of each column.

```python
#check dtypes
df.dtypes.unique()
```

We see that there are 3 types of data: int64, float64, and object.

```python
#count number of columns for each data type
for i in df.dtypes.unique():
    print("Number of", i, "columns: ", len(df.select_dtypes(include=i).columns))
```

```python
#check min, median, max, quartiles, mean and std for numerical columns (i.e. int64 & float64)
df.describe() # alternatively, df.describe(include = "number")
```

```python
#check number of unique values, top entries and frequencies for object columns
df.describe(include="object")
```

<h3> 2.2. Data Analysis & Visualization </h3>

Now, let's check the distribution of the variables in this data set.

```python
plt.figure(figsize=(8,6))
plt.style.use("seaborn")
y_dist = df["y"].value_counts()
plt.pie(x=y_dist, autopct="%.1f%%", labels = y_dist.keys(), textprops={'fontsize': 14})
```

Starting with the variable that we want to predict, we see that it is lopsided with "no"s account for around 89% while "yes"es account for the remaining 11%.

<h3> 2.2.1. Distribution and Correlations of Numeric Variables </h3>

```python
# plot histograms for each column
num_cols = df.select_dtypes(exclude = "object").columns
i = 1
plt.figure(figsize=(20,16))
plt.style.use("seaborn")
for column in num_cols:
  plt.subplot(4, 3, i)
  sns.distplot(df[column], kde = False, color="#000080")
  i+=1
```

Now, let's see how much correlation there is among the numeric variables.

```python
#check correlations among all input variables using a clustermap (heatmap + dendrogram)
plt.figure(figsize=(1,1))
sns.clustermap(df.select_dtypes(exclude = "object").corr(), annot=True, center = 0, cmap="vlag")
```

We see significant correlation between several numeric variables. For example, we see that euribor3m has a correlation of 0.97 with emp.var.rate while it has a correlation of 0.95 with nr.employed.

Also, we notice that variables _duration_, _campaign_, _age_, _cons.conf.idx_ do not have significant correlations with other variables while _euribor3m_, _cons.price.idx_, _nr.employed_, _emp.var.rate_ have significant correlations with each other.

_pdays_ and _previous_ have moderate correlations with some of the other variables.

These findings of significant correlation among some variables suggest the use of PCA to reduce dimensions of this dataset while retaining as much information as possible.


A special variable to note in this case is duration. The authors let us know that this variable is only known after a call is performed. Let's explore the relationship between y and duration.

```python
sns.boxplot(data=df, x="y", y="duration")
```

```python
#summary statistics for duration when y is yes
df_yes = df[df["y"] == "yes"]
df_yes["duration"].describe()
```

```python
#summary statistics for duration when y is no
df_no = df[df["y"] == "no"]
df_no["duration"].describe()
```

We see that duration has different mean and standard deviations depending on whether y is yes or no.
We can also see that the difference is significant using t-test. As seen below, p-value is 0.0.

```python
tStat, pValue =  stats.ttest_ind(df_yes["duration"], df_no["duration"])
print("P-Value:{0} T-Statistic:{1}".format(pValue,tStat))
```

<h3> 2.2.2. Distribution of Categorical Variables </h3>

```python
i=1
plt.figure(figsize=(20,16))
plt.style.use("seaborn")
cat_cols = df.drop("y", axis=1).select_dtypes(include = "object").columns
for column in cat_cols:
    plt.subplot(4, 3, i)
    ax = sns.countplot(y=column,
                       hue="y",
                       orient="h",
                       data=df)
    i+=1
    ax.set_yticklabels(ax.get_yticklabels(), rotation=30, fontsize = 12)
plt.show()
```
